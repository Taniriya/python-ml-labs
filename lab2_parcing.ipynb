Выбрать новостной сайт (или любой другой, критерий - должен быть список однотипных данных). Получить и при помощи BeautifulSoup распарсить сайт, на выход должен получиться список (или DataFrame) с новостями. Входной параметр - дата, начиная с которой производится парсинг, либо количество новостей, которые необходимо распарсить.
Доп. баллы - кроме заголовка новости получать ее текст и тоже добавлять в DataFrame.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd


df = pd.DataFrame(columns=('Headings', 'Article Overview', 'Text'))
url = 'https://www.rbc.ru/'
# Выполняем запрос на страницу и получаем её код
response = requests.get(url)
bs = BeautifulSoup(response.text, "html.parser")
# Получаем заголовки
news_headings = bs.find_all('a', class_="main__feed__link js-yandex-counter js-rbcmetrika-counter js-visited")
# Проходимся по каждому из них
for i in range(len(news_headings)):
  # Получаем текст каждого заголовка
  news_heading = news_headings[i].find('span', class_='g-inline-text-badges__text').get_text().strip()
  # И ссылку на статью
  news_text_link = news_headings[i]['href']
  # Выполняем запрос на статью и получаем её код
  response = requests.get(news_text_link)
  news = BeautifulSoup(response.text, "html.parser")
  # Получаем краткое описание статьи
  try:
    news_overview = news.find('div', class_='article__text__overview').get_text().strip()
  except AttributeError:
    news_overview = 'None'
  news_text = ''
  # И текст
  try:
    news_texts = news.find('div', class_='article__text article__text_free').find_all('p')
    for i in range(len(news_texts)):
      news_text += news_texts[i].get_text() + ' '
  except AttributeError:
    news_text = f'Объёмная статья. Советую почитать по ссылке: {news_text_link}'
  # Обозначаем платные статьи
  if news_text.strip() == '':
    news_text = 'Платная статья'
  # Полученные данные заносим в таблицу
  df.loc[len(df) + 1] = [news_heading, news_overview, news_text]
# И выводим
df
```
